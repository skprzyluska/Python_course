```python
# Zadanie 1
import random

lista = []

for i in range(100):
    a = int(random.uniform(1, 100))
    lista.append(a)

for i in range(100):
    print(lista[i])
lista.reverse()
print("\n")
for i in range(100):
    print(lista[i])
```

```python
# Zadanie 2

temperatura = float(input("Temperature in C :"))

if temperatura < -30:
    print ("ostry mróz")
elif temperatura < 0:
    print ("mróz")
elif temperatura < 10:
    print("zimno")
elif temperatura < 15:
    print("rzeźko")
elif temperatura < 25:
    print("optymalnie")
else:
    print("upał...")
```

```python
# Zadanie 3

lista = input("Enter elements: ").split() 

wynik = 0

for i in range (len(lista)):
    lista[i] = float(lista[i])
    if lista[i] == int(lista[i]):
        wynik += int(lista[i])
print (wynik)
 
```
